### 2. Sales

```python
sales=pd.read_csv('FACT_SALES.csv')
sales
```

```python
sales.info()
```

```python
sales['WEEK']=sales['WEEK'].astype('string')
sales['ITEM_CODE']=sales['ITEM_CODE'].astype('string')
sales['REGION']=sales['REGION'].astype('string')
```

```python
sales.info()
```

```python
sales.isnull().sum()
```

```python
print(sales.duplicated().sum())
```

```python
# Se filtra el dataframe sin Región para ver qué filas se repiten
sales_duplicated=sales.loc[sales[['WEEK','ITEM_CODE','TOTAL_UNIT_SALES',
                       'TOTAL_VALUE_SALES','TOTAL_UNIT_AVG_WEEKLY_SALES']].duplicated()]
sales_duplicated
```

```python
# Duplicados del total de regiones
sales_region=sales_duplicated.loc[sales_duplicated['REGION']=='TOTAL AUTOS SCANNING MEXICO']
sales_region
```

```python
#Duplicados diferentes del total de regiones
sales_region1=sales_duplicated.loc[sales_duplicated['REGION']!='TOTAL AUTOS SCANNING MEXICO']
sales_region1
```

```python
#Se ordena el dataframe por Región para facilitar el borrado de filas
sales_orden=sales.sort_values(by='REGION')
```

```python
sales_orden
```

```python
# Eliminamos la columna Región para poder identificar las filas duplicadas
sales_orden_sin_region =sales_orden.drop(columns=['REGION'])
```

```python
sales_orden_sin_region
```

```python
#Se eliminan los últimos duplicados, se conservan los primeros
sales_df=sales_orden_sin_region.drop_duplicates(keep='first')
sales_df
```

```python
#Unir dataframes (Región) por índice predeterminado mediante el método Merge (similar a join de SQL)
sales_sales_df = sales_df.merge(sales_orden['REGION'], left_index=True, right_index=True, how='inner')
sales_sales_df 
```
